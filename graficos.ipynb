```python
#IMPORTANDO OS PACOTES
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings

```

```python
warnings.simplefilter(action='ignore', category=FutureWarning)
data = pd.read_excel('data/base_dados_modelo.xlsx')
data['date'] = pd.to_datetime(data['date'])

# Filter the data
data = data[data['date'] < pd.to_datetime("2019-12-01")]

# Select columns
cols_select = ['date', 'selic', 'acum_12_meses', 'meta_inflacao', 'banda',
               'lim_inf', 'lim_sup', 'meta_modificada',
               'JUROS_NATURAL', 'REER', 'IBC_BR', 'COMMODITIES', 'EXP_PIB',
               'EXP_INF', 'gap_inflacao',
               'LOSS_12']

def meta_plot():
    sns.set_theme()
    plt.figure(figsize=(8, 4))
    sns.lineplot(x='date', y='acum_12_meses', data=data, label='Inflação')
    sns.lineplot(x='date', y='meta_inflacao', data=data, label='Meta da Inflação', linestyle='dashed', color='red')
    sns.lineplot(x='date', y='meta_modificada', data=data, linestyle='solid', color='black', label='meta modificada')
    plt.title(None)
    plt.xlabel('Data')
    plt.ylabel('% a.a')
    plt.legend()
    plt.show()

def gap_plot():
    plt.figure(figsize=(8, 4))
    gap_cred = sns.lineplot(x='date', y='LOSS_12', data=data, label='Perda de Credibilidade', linestyle='solid', color='red')
    sns.lineplot(x='date', y='gap_inflacao', data=data, label='Gap de Inflação', linestyle='dashed')
    plt.axhline(y=0, linestyle='solid', color='black')
    plt.title(None)
    plt.xlabel(None)
    plt.ylabel(None)
    plt.legend()
    plt.show()
```

# Gráfico de inflação

Este gráfico plota o IPCA acumulado em 12 meses, a meta da inflação anunciada pelo BC e a meta de inflação modificada

```python
meta_plot()
```

Este gráfico plota o indicador Perda de Credibilidade e o Gap de inflação

```python
gap_plot()
```
